```python
from sklearn.ensemble import RandomForestRegressor

X, y = df[['enginesize']], df[['price']]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=777)
reg = RandomForestRegressor(n_estimators=100,# default 
                            criterion='mse', # default
                            random_state=777, 
                            n_jobs=-1)
reg.fit(X_train, y_train)

x_plot = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_plot = reg.predict(x_plot)
plt.scatter(X_train, y_train, label='Training points', alpha=.4)
plt.plot(x_plot, y_plot, 
         color='black', lw=3, linestyle='-')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

y_train_pred = reg.predict(X_train)
y_test_pred = reg.predict(X_test)

print('MSE(training): %.3f, MSE(testing): %.3f' %( 
    mean_squared_error(y_train, y_train_pred), 
    mean_squared_error(y_test, y_test_pred)))
print('MAE(training): %.3f, MAE(testing): %.3f' %( 
    mean_absolute_error(y_train, y_train_pred), 
    mean_absolute_error(y_test, y_test_pred)))
print('R^2(training): %.3f, R^2(testing): %.3f' %( 
    r2_score(y_train, y_train_pred), 
    r2_score(y_test, y_test_pred)))
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler

model3 = sm.OLS(y, X).fit()
df_resid3 = pd.DataFrame()
df_resid3['y_pred'] = model3.predict(X)
df_resid3['resid'] = model3.resid
df_resid3 = StandardScaler().fit_transform(df_resid3)
```

```python
from statsmodels.stats.stattools import durbin_watson

dw = durbin_watson(model3.resid)
print('dw: %.3f' % dw)

if 2 <= dw <= 4:
    print('誤差項獨立')
elif 0 <= dw < 2:
    print('誤差項不獨立')
else:
    print('計算錯誤')
```

```python
from scipy.stats import shapiro
from statsmodels.graphics.gofplots import qqplot

stat, p = shapiro(model3.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
    
#QQ圖
qqplot(model3.resid, line='s');
```
