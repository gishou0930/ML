# Part2. 敘述性統計分析

```python
df.describe()
```

```python
#將非數值的type編碼
fueltype_code = pd.factorize(df["fueltype"]) #['gas', 'diesel']
df["fueltype_code"] = fueltype_code[0]

aspiration_code = pd.factorize(df["aspiration"]) #['std', 'turbo']
df["aspiration_code"] = aspiration_code[0]

doornumber_code = pd.factorize(df["doornumber"]) #['two', 'four']
df["doornumber_code"] = doornumber_code[0]

carbody_code = pd.factorize(df["carbody"]) #['convertible', 'hatchback', 'sedan', 'wagon', 'hardtop']
df["carbody_code"] = carbody_code[0]

drivewheel_code = pd.factorize(df["drivewheel"]) #['rwd', 'fwd', '4wd']
df["drivewheel_code"] = drivewheel_code[0]

enginelocation_code = pd.factorize(df["enginelocation"]) #['front', 'rear']
df["enginelocation_code"] = enginelocation_code[0]

enginetype_code = pd.factorize(df["enginetype"]) #['dohc', 'ohcv', 'ohc', 'l', 'rotor', 'ohcf', 'dohcv']
df["enginetype_code"] = enginetype_code[0]

cylindernumber_code = pd.factorize(df["cylindernumber"]) #['four', 'six', 'five', 'three', 'twelve', 'two', 'eight']
df["cylindernumber_code"] = cylindernumber_code[0]

fuelsystem_code = pd.factorize(df["fuelsystem"]) #['mpfi', '2bbl', 'mfi', '1bbl', 'spfi', '4bbl', 'idi', 'spdi']
df["fuelsystem_code"] = fuelsystem_code[0]

df
```

```python
df.describe()
```

# Part3. 特徵相關性分析

```python
#畫熱力圖找與price相關係數高的類別
corr = df.loc[:, 'symboling':'fuelsystem_code'].corr()
sns.set(rc = {'figure.figsize':(20,20)}) #調整heatmap大小
ax = sns.heatmap(
    corr, vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    annot = True,
    annot_kws={"fontsize":15},
    fmt = ".2f"
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

# Part4. 資料分割與建置迴歸模型

```python
#找相關性最高的enginesize做迴歸
x, y = df.loc[:, ['enginesize']], df.loc[:, ['price']]

x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.3, 
                                                    random_state=777, 
                                                    )

lr = LinearRegression()
lr.fit(x_train, y_train)
print('w_1 =', lr.coef_[0])
print('w_0 =', lr.intercept_)

plt.scatter(x_train, y_train, facecolor='xkcd:azure', edgecolor='black', s=20)
plt.xlabel('enginesize', fontsize=14)
plt.ylabel("price", fontsize=14)
# 繪製迴歸線
n_x = np.linspace(x_train.min(), x_train.max(), 100)
n_y = lr.intercept_ + lr.coef_[0] * n_x
plt.plot(n_x, n_y, color='r', lw=3);
```

# Part5. 模型效能評估

```python
#預測
y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
```

```python
#MSE
print('MSE(training): %.3f, MSE(testing): %.3f' %( 
    mean_squared_error(y_train, y_train_pred), 
    mean_squared_error(y_test, y_test_pred)))
```

```python
#MAE
print('MAE(training): %.3f, MAE(testing): %.3f' %( 
    mean_absolute_error(y_train, y_train_pred), 
    mean_absolute_error(y_test, y_test_pred)))
```

```python
#R Square
print('R^2(training): %.3f, R^2(testing): %.3f' %( 
    r2_score(y_train, y_train_pred), 
    r2_score(y_test, y_test_pred)))
```

```python
model1 = sm.OLS(y, x).fit()

# Shapiro-Wilk 常態性檢定
stat, p = shapiro(model1.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
```

```python
#QQ圖
qqplot(model1.resid, line='s');
```

```python
#判斷獨立性
dw = durbin_watson(model1.resid)
print('dw: %.3f' % dw)

if 2 <= dw <= 4:
    print('誤差項獨立')
elif 0 <= dw < 2:
    print('誤差項不獨立')
else:
    print('計算錯誤')
```

```python
#判斷變異數同質性
df_resid = pd.DataFrame()
df_resid['y_pred'] = model1.predict(x)
df_resid['resid'] = model1.resid
df_resid = StandardScaler().fit_transform(df_resid)

kws = {'color':'red', 'lw':3}
sns.residplot(x=df_resid[:, 0], y=df_resid[:, 1], 
              lowess=True, line_kws=kws)
plt.xlabel('Predicted Values', fontsize=14)
plt.ylabel('Residual', fontsize=14)
```

# Part6. 預測結果分析

```python
y_pred = lr.predict(x_test)#對測試集數據，用predict函數預測

plt.plot(range(len(y_pred)),y_pred,'red', linewidth=2.5,label="predict data")
plt.plot(range(len(y_test)),y_test,'green',label="test data")
plt.legend(loc=2)
plt.show()#顯示預測值與測試值曲線
```

# Part7. 採用多元線性迴歸

```python
#做一個新的df，包含Price以及與Price高度正相關(>0.7)的類別出來
new_df = df[["price","carwidth","curbweight","enginesize","horsepower"]]
new_df
```

```python
X = new_df.loc[:, 'carwidth':'horsepower']   # 特徵
y = new_df['price']               # 目標類別
# 切割數據集
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=777, 
                                                    )
#標準化
scale = StandardScaler().fit(X_train)
X_train_std = scale.transform(X_train)
X_test_std = scale.transform(X_test)

# 觀看數據集、訓練與測試集的類別比例
df_count = pd.concat([y.value_counts(), 
                      y_train.value_counts(), 
                      y_test.value_counts()], axis=1)
df_count.columns = ['y', 'y_train', 'y_test']
df_count
```

```python
#MSE、MAE、R Square

lr2 = LinearRegression()
lr2.fit(X_train_std, y_train)
y_train_pred = lr2.predict(X_train_std)
y_test_pred = lr2.predict(X_test_std)
print('MSE(training): %.3f, MSE(testing): %.3f' %( 
    mean_squared_error(y_train, y_train_pred), 
    mean_squared_error(y_test, y_test_pred)))
print('MAE(training): %.3f, MAE(testing): %.3f' %( 
    mean_absolute_error(y_train, y_train_pred), 
    mean_absolute_error(y_test, y_test_pred)))
print('R^2(training): %.3f, R^2(testing): %.3f' %( 
    r2_score(y_train, y_train_pred), 
    r2_score(y_test, y_test_pred)))
```

```python
# Shapiro-Wilk 常態性檢定、QQ圖
model2 = sm.OLS(y, X).fit()
stat, p = shapiro(model2.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
    
#QQ圖
qqplot(model2.resid, line='s');
```

```python
#DW檢測
dw = durbin_watson(model2.resid)
print('dw: %.3f' % dw)

if 2 <= dw <= 4:
    print('誤差項獨立')
elif 0 <= dw < 2:
    print('誤差項不獨立')
else:
    print('計算錯誤')
```

```python
#殘差圖
df_resid2 = pd.DataFrame()
df_resid2['y_pred'] = model2.predict(X)
df_resid2['resid'] = model2.resid
df_resid2 = StandardScaler().fit_transform(df_resid2)

kws = {'color':'red', 'lw':3}
sns.residplot(x=df_resid2[:, 0], y=df_resid2[:, 1], 
              lowess=True, line_kws=kws)
plt.xlabel('Predicted Values (standardization)', fontsize=14)
plt.ylabel('Residual (standardization)', fontsize=14)
```

```python
#預測

plt.plot(range(len(y_test_pred)),y_test_pred,'red', linewidth=2.5,label="predict data")
plt.plot(range(len(y_test)),y_test,'green',label="test data")
plt.legend(loc=2)
plt.show()#顯示預測值與測試值曲線
```

# Part8. 採用隨機森林迴歸

```python
X, y = df[['enginesize']], df[['price']]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=777)
reg = RandomForestRegressor(n_estimators=100,# default 
                            criterion='mse', # default
                            random_state=0, 
                            n_jobs=-1)
reg.fit(X_train, y_train)

x_plot = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_plot = reg.predict(x_plot)
plt.scatter(X_train, y_train, label='Training points', alpha=.4)
plt.plot(x_plot, y_plot, 
         color='black', lw=3, linestyle='-')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.legend()
```

```python
#MSE、MAE、R Square
y_train_pred = reg.predict(X_train)
y_test_pred = reg.predict(X_test)

print('MSE(training): %.3f, MSE(testing): %.3f' %( 
    mean_squared_error(y_train, y_train_pred), 
    mean_squared_error(y_test, y_test_pred)))
print('MAE(training): %.3f, MAE(testing): %.3f' %( 
    mean_absolute_error(y_train, y_train_pred), 
    mean_absolute_error(y_test, y_test_pred)))
print('R^2(training): %.3f, R^2(testing): %.3f' %( 
    r2_score(y_train, y_train_pred), 
    r2_score(y_test, y_test_pred)))
```

```python
model3 = sm.OLS(y, X).fit()
df_resid3 = pd.DataFrame()
df_resid3['y_pred'] = model3.predict(X)
df_resid3['resid'] = model3.resid
df_resid3 = StandardScaler().fit_transform(df_resid3)
```

```python
#SW檢測、QQ圖
stat, p = shapiro(model3.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
    
#QQ圖
qqplot(model3.resid, line='s');
```

```python
#DW檢測
dw = durbin_watson(model3.resid)
print('dw: %.3f' % dw)

if 2 <= dw <= 4:
    print('誤差項獨立')
elif 0 <= dw < 2:
    print('誤差項不獨立')
else:
    print('計算錯誤')
```

```python
#殘差圖
sns.residplot(x=df_resid3[:, 0], y=df_resid3[:, 1], lowess=True,  
              line_kws=kws)

plt.legend()
plt.xlabel('Predicted')
plt.ylabel('Residual')
```

```python
#顯示預測值與測試值曲線
plt.plot(range(len(y_test_pred)),y_test_pred,'red', linewidth=2.5,label="predict data")
plt.plot(range(len(y_test)),y_test,'green',label="test data")
plt.legend(loc=2)
plt.show()
```
