```python
from sklearn.covariance import EllipticEnvelope
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

def outlier_idx(x):
    q1, q3 = np.percentile(x, [25, 75])
    IQR = q3 - q1
    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR
    return np.where((x < lower_bound) | (x > upper_bound))

# 找出離群值
outlier = ["enginesize","boreratio","compressionratio","horsepower",]
idx = outlier_idx(df['horsepower'])[0]
df['Outlier'] = [1 if i in idx else 0 for i in df.index]
df.take([0] + idx)
```

```python
#將非數值的type編碼
fueltype_code = pd.factorize(df["fueltype"]) #['gas', 'diesel']
df["fueltype_code"] = fueltype_code[0]

aspiration_code = pd.factorize(df["aspiration"]) #['std', 'turbo']
df["aspiration_code"] = aspiration_code[0]

doornumber_code = pd.factorize(df["doornumber"]) #['two', 'four']
df["doornumber_code"] = doornumber_code[0]

carbody_code = pd.factorize(df["carbody"]) #['convertible', 'hatchback', 'sedan', 'wagon', 'hardtop']
df["carbody_code"] = carbody_code[0]

drivewheel_code = pd.factorize(df["drivewheel"]) #['rwd', 'fwd', '4wd']
df["drivewheel_code"] = drivewheel_code[0]

enginelocation_code = pd.factorize(df["enginelocation"]) #['front', 'rear']
df["enginelocation_code"] = enginelocation_code[0]

enginetype_code = pd.factorize(df["enginetype"]) #['dohc', 'ohcv', 'ohc', 'l', 'rotor', 'ohcf', 'dohcv']
df["enginetype_code"] = enginetype_code[0]

cylindernumber_code = pd.factorize(df["cylindernumber"]) #['four', 'six', 'five', 'three', 'twelve', 'two', 'eight']
df["cylindernumber_code"] = cylindernumber_code[0]

fuelsystem_code = pd.factorize(df["fuelsystem"]) #['mpfi', '2bbl', 'mfi', '1bbl', 'spfi', '4bbl', 'idi', 'spdi']
df["fuelsystem_code"] = fuelsystem_code[0]

df
```

# Part4.資料分割與建置迴歸模型

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = new_df.loc[:, 'carwidth':'horsepower']   # 特徵
y = new_df['price']               # 目標類別
# 切割數據集
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=777, 
                                                    )
#標準化
scale = StandardScaler().fit(X_train)
X_train_std = scale.transform(X_train)
X_test_std = scale.transform(X_test)

# 觀看數據集、訓練與測試集的類別比例
df_count = pd.concat([y.value_counts(), 
                      y_train.value_counts(), 
                      y_test.value_counts()], axis=1)
df_count.columns = ['y', 'y_train', 'y_test']
df_count
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train_std, y_train)
print('[w_1, w_2] =', lr.coef_[0])
print('w_0 =', lr.intercept_)
```

```python
a  = lr.intercept_#截距
b = lr.coef_#迴歸係數
print("擬合參數:截距",a,",迴歸係數：",b)
```

```python
#顯示線性方程，並限制參數的小數位爲兩位
print("最佳擬合線: Y = ",round(a,2),"+",round(b[0],2),"* X1 + ",round(b[1],2),"* X2",round(b[2],2),"* X3",round(b[3],2),"* X4")

y_pred = lr.predict(X_test_std)#對測試集數據，用predict函數預測

plt.plot(range(len(y_pred)),y_pred,'red', linewidth=2.5,label="predict data")
plt.plot(range(len(y_test)),y_test,'green',label="test data")
plt.legend(loc=2)
plt.show()#顯示預測值與測試值曲線
```

```python
from sklearn.metrics import mean_squared_error

y_train_pred = lr.predict(X_train_std)
y_test_pred = lr.predict(X_test_std)
print('MSE(training): %.3f, MSE(testing): %.3f' %( 
    mean_squared_error(y_train, y_train_pred), 
    mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.metrics import mean_absolute_error

print('MAE(training): %.3f, MAE(testing): %.3f' %( 
    mean_absolute_error(y_train, y_train_pred), 
    mean_absolute_error(y_test, y_test_pred)))
```

```python
from sklearn.metrics import r2_score

print('R^2(training): %.3f, R^2(testing): %.3f' %( 
    r2_score(y_train, y_train_pred), 
    r2_score(y_test, y_test_pred)))
```

```python
import statsmodels.api as sm

model = sm.OLS(y, X)
result = model.fit()
print('迴歸係數：', result.params)
result.summary()
```

```python
import statsmodels.api as sm
from scipy.stats import shapiro

# Shapiro-Wilk 常態性檢定
stat, p = shapiro(result.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
```

```python
from scipy.stats import kstest

# Kolmogorov-Smirnov 常態性檢定
stat, p = kstest(result.resid, 'norm')
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
```

```python
from scipy.stats import normaltest

# D’Agostino-Pearson omnibus 常態性檢定
stat, p = normaltest(result.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
```

```python
from scipy.stats import jarque_bera

# Jarque-Bera 常態性檢定
stat, p = jarque_bera(result.resid)
print('Statistics: %.3f, p-value: %.3f' % (stat, p))
alpha = 0.05

if p > alpha:
    print('看起來是常態分布（無法拒絕H0）')
else:
    print('看起來不是常態分布（拒絕H0）')
```

```python
from statsmodels.graphics.gofplots import qqplot

qqplot(result.resid, line='s');
```

```python
from statsmodels.stats.stattools import durbin_watson

dw = durbin_watson(result.resid)
print('dw: %.3f' % dw)

if 2 <= dw <= 4:
    print('誤差項獨立')
elif 0 <= dw < 2:
    print('誤差項不獨立')
else:
    print('計算錯誤')
```
